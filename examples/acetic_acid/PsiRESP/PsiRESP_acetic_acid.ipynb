````python
import numpy as np
import psiresp

# ... Assuming you have run your job: charges = job.run() ...
# Let's say your molecule object is called `molecule`

rmse_list = []

# Loop through every conformer generated or used for the molecule
for conformer in molecule.conformers:
    # Loop through each rigid-body orientation of that conformer
    for orientation in conformer.orientations:
        
        # 1. Get the QM calculated coordinates and ESP grid values
        grid = orientation.grid              # Shape: (N, 3) in Bohr
        qm_esp = orientation.esp             # Shape: (N,) in atomic units
        
        # 2. Get the Cartesian coordinates of the atoms for this orientation
        # (Converted to Bohr to match the grid units)
        atom_coords = orientation.coordinates 
        
        # 3. Get the final derived charges
        # (You can also fetch molecule.stage_1_restrained_charges, etc.)
        charges = molecule.charges
        
        # 4. Calculate the classical Coulombic potential generated by the charges at each grid point
        calc_esp = np.zeros_like(qm_esp)
        for i, grid_point in enumerate(grid):
            # Distance from this grid point to all atoms
            distances = np.linalg.norm(atom_coords - grid_point, axis=1)
            # Coulombic potential: sum(q_j / r_ij)
            calc_esp[i] = np.sum(charges / distances)
            
        # 5. Compute the RMSE for this specific orientation
        rmse = np.sqrt(np.mean((qm_esp - calc_esp) ** 2))
        rmse_list.append(rmse)

        print(f"Orientation RMSE: {rmse:.6f} a.u.")

        calc_esp = np.zeros_like(qm_esp)
        for i, grid_point in enumerate(grid):
            distances = np.linalg.norm(atom_coords - grid_point, axis=1)
            calc_esp[i] = np.sum(charges / distances)
            
        # 2. Compute absolute differences
        residuals_sq = (qm_esp - calc_esp) ** 2
        qm_esp_sq = qm_esp ** 2
        
        # 3. Compute RRMSE
        rrmse = np.sqrt(np.sum(residuals_sq) / np.sum(qm_esp_sq))
        
        # Alternatively, expressed as a percentage:
        rrmse_percentage = rrmse * 100
        
        print(f"Orientation RRMSE: {rrmse:.4f} ({rrmse_percentage:.2f}%)")

              
print(f"Average System RMSE across all orientations: {np.mean(rmse_list):.6f} a.u.")
````
