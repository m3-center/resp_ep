# Adding Extra Point Centers to XYZ-formatted Files
---

```python
import os

from resp_ep import extras
```

## Adding Extra Points Centers That Model Electron Lone Pairs

### Models $sp^3$ Orbitals
- For example, alcohol oxygen atom

```python
extras.add_lone_pairs(molecule="acetic_acid",
                      lp_specs=[{"center": 6, "neighbors": [5, 7], "plane_atom": None}],
                      lp_distance=0.35,
                      lp_angle_deg=109.5,
                      mode="out_of_plane",
                      work_dir="./")
```

### Models $sp^2$ Orbitals
- For example, carbonyl oxygen atom

- We will add them to the structure generated above, so that we have both sets of electron lone pairs modelled.

```python
extras.add_lone_pairs(molecule="acetic_acid_x",
                      lp_specs=[{"center": 8, "neighbors": [5], "plane_atom": 6}],
                      lp_distance=0.35,
                      lp_angle_deg=120.0,
                      mode="in_plane",
                      work_dir="./")
```

```python
os.rename("acetic_acid_x_x.xyz", "acetic_acid_x.xyz")
```
